```python
# part A
# import the core layers:
# "Sequential" models let usv define a stack of neural network layers
from keras.models import Sequential

# import the core layers:
from keras.layers import Dense, Dropout, Activation, Flatten

import numpy as np
from keras.utils import np_utils

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV



```

```python
#part B
from keras.datasets import mnist

# By default, the first 60k of MNIST has been defined as training and the rest as testing set: 
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# reshape to a single row of 784
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)

# simply scale the features to the range of [0,1]:
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255
```

```python
#part C 

#perform OneHotEncoding
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)

# Label after OneHotEncoding:
print (y_train.shape)
print (y_train[:10,:])
```

```python
def model_creator():
    #define:
    model = Sequential()
    input_size = 784
    hidden_neurons = 100
    out_size = 10
    #design the structure:
    ## Designing the NN Structure:

    # -----------------------------------------
    # first layer: input layer
    # Input layer does not do any processing, so no need to define the input layer in this problem.

    # -----------------------------------------
    # second layer: hidden layer:
    model.add(Dense(hidden_neurons, input_dim = input_size))  # Nuerons
    model.add(Activation('sigmoid')) # Activation

    # -----------------------------------------
    # third layer: output layer:
    model.add(Dense(out_size, input_dim = hidden_neurons))  # Nuerons
    model.add(Activation('softmax')) # Activation
    
    #compile
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    #return
    return model
```

```python
#part E 
seed = 2
np.random.seed(seed)
```

```python
#part F
model = KerasClassifier(build_fn = model_creator, verbose=2)
```

```python
#part G
param_grid=dict(batch_size = [30 , 50 , 100 ] , epochs = [10 , 15 , 20])

grid = GridSearchCV(model, param_grid, cv=10, scoring='neg_log_loss')

grid=grid.fit(X_train,y_train)
# view the results:
#print(grid.cv_results_)

```

```python
print(grid.best_score_)
print(grid.best_params_)
```

```python
# part H

# Based on your results, what is the best batch_size and epochs?
# {'batch_size': 50, 'epochs': 20}

# What is the prediction accuracy on the testing set?
# {'accuracy of best_estimator model': 0.9773}

model= grid.best_estimator_
y_pred = model.predict(X_test,verbose=1)
#print(y_pred)

score = model.score(X_test, y_test)  # this .score() method makes
#                                      -it easier to get accuracy!!!!

# ---NOTE ===> since we 1HotEncoded the data -
#----      we have to do the same to results!!
#-----     else we have to do OneHotEncode to the prdeiction(s).

predict_1hot_encode = np_utils.to_categorical(y_pred, 10)
print(score)
#print(y_test)
#print(predict_1hot_encode)


from sklearn.metrics import accuracy_score
accuracy_results={}
accuracy_results['accuracy of best_estimator model']=accuracy_score(y_test,predict_1hot_encode)
print(accuracy_results)

```
